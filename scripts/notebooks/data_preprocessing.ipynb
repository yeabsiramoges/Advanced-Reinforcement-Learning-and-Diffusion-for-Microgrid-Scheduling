```python
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
```

```python
# Seed
rand_seed = 1234

# Load training data
dataset = pd.read_csv("../../data/rye/train.csv", index_col=0, parse_dates=True)
dataset.head()
```

```python
# Missing values
assert dataset[dataset.photovoltaic_production >= 0].size == dataset.size

dataset = dataset.dropna(how="all")
```

```python
# Feature extraction
pca = PCA(2).fit(dataset)

print(pca.transform(dataset))
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```
